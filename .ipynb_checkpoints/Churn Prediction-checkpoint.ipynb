### 3 Preprocessing

```python
# replacing the values
d = {'Yes' : True, 'No' : False}
df = df.replace({'International plan' : d,'Voice mail plan' :d})
df.head()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # object created
```

```python
# converting categorical values to numerical values
col = (['International plan','Voice mail plan','Churn'])
for i in col:
    df[i] = le.fit_transform(df[i])
df.head()
```

```python
# feture selection -> selecting input and output variables
x = df[['Account length', 'Area code', 'International plan',
       'Voice mail plan', 'Number vmail messages', 'Total day minutes',
       'Total day calls', 'Total day charge', 'Total eve minutes',
       'Total eve calls', 'Total eve charge', 'Total night minutes',
       'Total night calls', 'Total night charge', 'Total intl minutes',
       'Total intl calls', 'Total intl charge', 'Customer service calls']]
y = df['Churn']
```

### 4 split train and test data

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
print('shape of x_train: ',x_train.shape)
print('shape of x_test: ',x_test.shape)
print('shape of y_train: ',y_train.shape)
print('shape of y_test: ',y_test.shape)
```

### 5 model building

```python
# knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

### 6 Model training

```python
knn.fit(x_train,y_train) # training
```

### 7 Model testing

```python
pred_y1 = knn.predict(x_test) # testing
pred_y1
```

### 8 Evaluation metrics

```python
#evaluation metrics
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
#accuracy
accuracy1 = accuracy_score(y_test,pred_y1)
#f1_score
f1_score1 = f1_score(y_test,pred_y1)
#precision
precision1 = precision_score(y_test,pred_y1)
#recall
recall1 = recall_score(y_test,pred_y1)

print('accuracy: ',accuracy1)
print('f1_score: ',f1_score1)
print('precision: ',precision1)
print('recall: ',recall1)
```

### 9 predicting unseen data

```python
val = np.array([200,436,1,0,0,182,99,39.2,330.1,85,29.89,215.8,90,9.71,8.7,4,2.35,1])
val = val.reshape(1,18)
val
```

```python
output = knn.predict(val)
d = {0:'Bad',1:"Good"}
print(d[output[0]])
```

### Plotting graphs

```python
acc_d = [accuracy1*100,accuracy2*100,accuracy3*100] 
#considering percentages
x = np.array([1,2,3]) # array of indices
models=["KNN","Decision tree","Naive bayes"]
data = pd.DataFrame({'Models':models,"Accuracies":acc_d})
# datafram
data['Accuracies'] = data['Accuracies'].round(2)
plt.bar(data['Models'],data['Accuracies'])
# bar plotted is plotted
for i, v in enumerate(data['Accuracies']):
    # i is index and v is value, enumerate is used to loop over i and v simultaneously
    plt.text(i, v + 0.01, str(v), ha='center',va='bottom')
    # text is added at (i,v+0.01) and number is coverted to string using str()
    # ha - horizontal alignment , va - vertical alignment
    # bottom - Aligns the text as bottom of text is at the specified position.
plt.xlabel("Models")
plt.ylabel("Accuarcies")
plt.show()
```
