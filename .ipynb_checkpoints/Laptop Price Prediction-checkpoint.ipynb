### 3) Data preprocessing

```python
df.isna().sum() #sum of null values present in each column
```

```python
# there are no wrong data and no null values
df.duplicated() #checks whether there are any duplicates present in the dataset
```

```python
df.duplicated().sum() # gives the sum value of no.of duplicates in dataset
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.shape
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # object created
```

```python
df.columns # displays all columns
```

```python
# converting categorical values to numerical values
c = (['brand','processor_brand', 'processor_name', 'processor_gnrtn',
       'ram_gb', 'ram_type', 'ssd', 'hdd', 'os', 'os_bit', 'graphic_card_gb',
       'weight', 'warranty', 'Touchscreen', 'msoffice','rating'])
for i in c:
    df[i] = le.fit_transform(df[i])

df
```

```python
# feture selection -> selecting input and output variables
x = df[['brand', 'processor_brand', 'processor_name', 'processor_gnrtn',
       'ram_gb', 'ram_type', 'ssd', 'hdd', 'os', 'os_bit', 'graphic_card_gb',
       'weight', 'warranty', 'Touchscreen', 'msoffice', 'rating',
       'Number of Ratings', 'Number of Reviews']] #input features
y = df['Price']
```

### 4) Splitting dataset into training and testing data

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
#test data -> 33% train data-> 67%
print('shape of x_train: ',x_train.shape)
print('shape of x_test: ',x_test.shape)
print('shape of y_train: ',y_train.shape)
print('shape of y_test: ',y_test.shape)
```

### 5) model building

```python
#importing the model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

### 6) Model training

```python
lr.fit(x_train,y_train)
```

### 7) testing the model

```python
y_pred = lr.predict(x_test) # for predicting the values
y_pred
```

### 8) Evaluation Metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_test, y_pred) #mean Square error
mae = mean_absolute_error(y_test, y_pred) # mean absolute error
r2 = r2_score(y_test, y_pred) #r2 score

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

### 9) predicting unseen data

```python
s = np.array([5,1,1,0,2,1,0,1,2,1,0,0,3,0,0,2,65,5])
s = s.reshape(1,18)
```

```python
output = lr.predict(s)
print(output[0])
```
