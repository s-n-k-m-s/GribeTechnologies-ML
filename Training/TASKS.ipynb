## numpy package

```python
import numpy as np
```

```python
# 2-dimensional array creation
arr = np.array([[1,2,3,4],[2,3,4,5],[3,4,5,6]])
print(arr)
```

```python
# array manipulation
re_arr = arr.reshape(4,3) #reshaping the 3x4 array to 4x3 array
re_arr
```

```python
re_arr[:,2] 
#slicing concept is used to retrieve the data, here we are displaying 3nd column
```

```python
# mathematical operations
arr2 =  np.array([[1.0,2.8,3.2,4],[1.4,3,5.7,7.8],[3.2,4,5.1,6]])
arr2 #creating another two-dimensional array
```

```python
summ = np.add(arr,arr2) # sum of two arrays
summ
```

```python
sub = np.subtract(arr,arr2) #difference of two arrays
sub
```

```python
mul = np.multiply(arr,arr2) #product of two arrays
mul
```

```python
div = np.divide(arr,arr2) #division of two arrays
div
```

```python
# Statistical analysis
np.mean(arr) #mean of the array arr
```

```python
np.std(arr) #standard deviation of array arr
```

```python
np.min(re_arr) #minimum value in the array re_arr
```

```python
np.max(re_arr) # maximum value in the array re_arr
```

## matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dataset2.csv')
df
```

```python
#line graph
plt.plot(df['item'],df['cost'])
plt.xlabel('items')
plt.ylabel('cost of the items')
plt.title('graph b/w items and cost')
plt.show()
```

```python
#bar graph
plt.bar(df['item'],df['no.of.items'])
plt.xlabel('items')
plt.ylabel('number of the items')
plt.title('items and number of items purchased')
plt.show()
```

```python
#scatter plot
plt.scatter(df['item'],df['cost'])
plt.xlabel('items')
plt.ylabel('cost of the items')
plt.title('graph b/w items and cost')
plt.show()
```

```python
#histogram
plt.hist(df['cost'],bins=10,edgecolor='red')
plt.xlabel('cost')
plt.ylabel('frequency')
plt.title('Histogram')
#here we are having 3 values which are greater than equal to 120
#1 value above 120 and below 540 
# 1 value above 540 and below 780
```

```python
#pie chart
plt.pie(df['cost'],labels=df['item'],autopct='%.2f%%')
plt.title('Pie chart')
plt.show()
```
